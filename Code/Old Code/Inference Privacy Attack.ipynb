***

Function to perform the inference attack.

```python
syn = mostly_sXs[0]
```

```python
syn_tree = cKDTree(syn[["latitude", "longitude"]])
```

```python
train_tree = cKDTree(train_data[["latitude", "longitude"]])
```

```python
neighbors = train_tree.query_ball_tree(syn_tree, r=0.01, p=2.0)
```

```python
data_prob_death = []
for i in range(train_data.shape[0]):
    data_prob_death.append(np.mean(syn.iloc[neighbors[i],:].state))
```

```python
data_prob_death = [0 if np.isnan(x) else x for x in data_prob_death]
```

```python
cutoffs = [0.05, 0.1, 0.15, 0.2, 0.3, 0.4, 0.5]
```

```python
predicted_state = pd.Series([0 if x <= 0.1 else 1 for x in data_prob_death], name="predicted_state")
```

```python
train_data = pd.concat([train_data, predicted_state], axis=1)
```

```python
pos_state = train_data.loc[train_data.state==1,:]
```

```python
np.mean(pos_state.state == pos_state.predicted_state)
```
